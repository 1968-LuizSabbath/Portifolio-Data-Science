```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
sales = pd.read_csv('/content/drive/MyDrive/supermarket_sales.csv')
df = pd.DataFrame(sales)
df
```

```python
#Exibir as primeiras linhas do Dataset
print(df.head())

#verificar as informações gerais do dataset(colunas, tipos, valores nulos)
print(df.info())

#Verificar se há valores nulos em cada coluna
print(df.isnull().sum())

#Verificar se há linhas duplicadas
print(df.duplicated().sum)
```

```python
#Remover valores nulos(se houver)
df = df.dropna()

#Converter a coluna de data para o formato datetime
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

#Cia colunas para o Mês e o Ano
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

#Exibir as primeiras linhas para verificar as mudanças
df.head()
```

```python
#Começar a analise descritiva dos dados.
#Iniciar fazendo algumas perguntas basicas

#QUAL O TOTAL DE VENDAS DO MÊS?
vendas_por_mes = df.groupby('Month')['Total'].sum()

#Exibir Resultado
print(vendas_por_mes)
```

```python
#QUAIS AS CATEGORIAS DE PRODUTOS MAIS VENDIDAS?
#Agrupar por linha de produto e somar o total de vendas
venda_por_categoria = df.groupby('Product line')['Total'].sum()

#Exibir Resultado
print(venda_por_categoria)
```

```python
#QUAL O TICKET MEDIO (VALOR MEDIO POR VENDA)
#Calcular o ticket médio
ticket_medio = df['Total'].sum() / len(df)

#Exibir Resultado
print(f'Ticket médio:{ticket_medio:.2f}')
```

```python
#Gerar grafico para padrões de venda
import matplotlib.pyplot as plt

vendas_por_mes.plot(kind='bar', color='skyblue')
plt.title('Total de Vendas por Mês')
plt.xlabel('Mês')
plt.ylabel('Total de Vendas')
plt.show()
```

```python
#Plotar um grafico para vendas por categoria
venda_por_categoria.plot(kind='bar', color='orange')
plt.title('Total de Vendas por Categoria de Produto')
plt.xlabel('Categoria de Produto')
plt.ylabel('Total de Vendas')
plt.show()
```

```python
#Analisando as vendas por método de pagamento

#Agrupar por metodo de pagamento e somar as vendas
vendas_por_pagamento = df.groupby('Payment')['Total'].sum()

#Exibir Resultado
print(vendas_por_pagamento)

#Visualizar com um grafico de barras
vendas_por_pagamento.plot(kind='bar' , color='green')
plt.title('Total de Vendas por Método de Pagamento')
plt.xlabel('Método de Pagamento')
plt.ylabel('Total de Vendas')
plt.show()
```

```python
#Vendas por Cidade

#Agrupar por cidade e somar o total de vendas
vendas_por_cidade = df.groupby('City')['Total'].sum()

#Exibir Resultado
print(vendas_por_cidade)

#Visualizar as vendas por cidade
vendas_por_cidade.plot(kind='bar' , color='purple')
plt.title('Total de Vendas por Cidade')
plt.xlabel('Cidade')
plt.ylabel('Total de Vendas')
```

```python
#Analise de venda por genero e somar as vendas
vendas_por_genero = df.groupby('Gender')['Total'].sum()

#Exibir Resultado
print(vendas_por_genero)
#Exibir o resultado
print('vendas_por_genero')

#Visualizar as vendas por genero com grafico
vendas_por_genero.plot(kind='bar' , color='red')
plt.title('Total de Vendas por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Total de Vendas')
plt.show()
```
